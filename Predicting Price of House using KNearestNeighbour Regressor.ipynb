# 3. Model Training and testing

Importing required libraries for the model

```python
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
```

Splitting feature and target

```python
x = df.drop(['Price'],axis = 1)
x
```

```python
y = df['Price']
y
```

```python
# Splitting data for training and testing

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
# using the StandardScaler is to ensure that all features are on a similar scale

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# Training the Nearest Neighbors Regressor and fitting the regressor model with training data

knr = KNeighborsRegressor(n_neighbors=2)
knr.fit(x_train,y_train)
```

```python
y_pred = knr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(knr.score(x_train,y_train))
print(knr.score(x_test,y_test))
```

# 4. Finding a better value of k

```python
k_value= []

for k in range(2,11):
    knr_model = KNeighborsRegressor(n_neighbors=k)
    knr_model.fit(x_train,y_train)
    y_predict = knr_model.predict(x_test)
    k_value.append({k,knr_model.score(x_train,y_train),knr_model.score(x_test,y_test)})
```
